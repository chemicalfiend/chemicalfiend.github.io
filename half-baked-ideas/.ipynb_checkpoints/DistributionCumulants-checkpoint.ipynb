```julia
using LinearAlgebra
using Plots
using FFTW
```

```julia
cums = [40.0, 100, 5000, 0.0];
```

```julia
function characteristic_function(x; cumz=cums)
    s = 0. + 0.0im
    for (n, cum) in enumerate(cumz)
        if (n < 20)
            s += cum * ((x*1.0im)^n) / factorial(n)
        else
            # Use stirling's approximation if n > 20
            s += cum * ((x*1.0im)^n) / (n^n * exp(n))
        end
    end
    return exp(s)
end
```

```julia
characteristic_function(0.02)
```

```julia
t = -1:0.1:1
plot(t, real.(characteristic_function.(t)), title="Real part of characteristic function", ylabel="phi(x)")
```

```julia
x = -50:0.01:50
ϕ = characteristic_function.(x)
F = fftshift(fft(ϕ)) # Fourier transform of characteristic function gives us the distribution
P = F;
```

```julia
plot(x, real.(abs.(P)), title="Probability Distribution", ylabel="P(x)")
```

```julia
# Smoothing via moving averages

binsize = 150

K = abs.(real.(P))

mov = []
xaxis = []
for i in 1:(length(enumerate(K))-binsize)
    sum = 0
    for j in 1:binsize
        sum += K[i+j-1]
    end
    push!(mov, sum/binsize)
    push!(xaxis, x[i])
end
```

```julia
plot(xaxis, mov, title="Probability Distribution Smoothed")
```

```julia
# Verifying that our distribution is right

mean = sum([xi * abs.(real.(mov[i])) for (i, xi) in enumerate(xaxis)])
variance = sum([xi^2 * abs.(real.(mov[i])) for (i, xi) in enumerate(xaxis)]) - (mean^2)

println(mean)
println(variance)
```
